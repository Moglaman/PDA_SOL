```python
import pandas as pd
import numpy as np
import requests,zipfile,io 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
x = np.linspace(0, 50, 20)
y = x*3
 
# plot the graph
plt.plot(x, y)
 
plt.xlim(0, 100)
plt.ylim(0, 200)  
plt.title('Отрисовки линии')
plt.ylabel('Y ось')   
plt.xlabel('X ось')   
plt.show()
```

```python
names = ['Java','Python','PHP','JavaScript','C#','C++']
rating = [22.2,	17.6,8.8,8,7.7,6.7]
plt.bar(names,rating)
plt.ylabel('Рейтинг')
plt.xlabel('Языки программирования')
plt.title('Рейтинг языков программирования')
plt.xticks(rotation = 90)
```

```python
data = {'США':	46,
'Великобритания':	27,
'Китай':	26,
'Россия':	19,
'Германия':	17
}

keys_data = []
values_data = []

for x, y in data.items():
    keys_data.append(x)
    values_data.append(y)
plt.pie(values_data, labels=keys_data,autopct='%1.1f%%')
plt.title('Пять самых успешных стран на летних Олимпийских играх 2016 года.')
plt.show()
```
